```python
#test the loader
data=pd.read_csv("../data/twosides_cleaned.csv",converters={
    'Side Effect Name':pd.eval
})
```

```python
x1s=data['X1']
x2s=data['X2']
labels=data['Side Effect Name']

zipped=list(zip(x1s,x2s,labels))
```

```python
from torch_geometric.loader import DataLoader
from molecule_dataloader import get_graphs

zipped=zipped[:10]
dataset=get_graphs(zipped)
```

```python
params={
    'batch_size':1,
    'shuffle':True,
    'num_workers':0
}
loader=DataLoader(dataset,**params,follow_batch=['x_s','x_t'])
```
